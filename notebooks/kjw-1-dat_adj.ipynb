```python
# 결측치 제거
import numpy as np
import pandas as pd

# ==============================
# 1. 파일 경로
# ==============================
input_csv_path = 'C:/potenup3/pj02_Daangn-marke/data/csv/merged_dedup_siglip2_labeled.csv'  # 원본 CSV 경로
output_csv_path = (
    'C:/potenup3/pj02_Daangn-marke/data/csv/merged_dedup_siglip2_labeled_clean.csv'
)

# ==============================
# 2. 데이터 로드
# ==============================
df = pd.read_csv(input_csv_path, low_memory=False)

print('원본 shape:', df.shape)

# ==============================
# 3. price 숫자형 변환
# - 문자열일 가능성 대비
# ==============================
df['price'] = (
    df['price']
    .astype(str)
    .str.replace(r'[^\d\.\-]', '', regex=True)
    .replace('', np.nan)
)
df['price'] = pd.to_numeric(df['price'], errors='coerce')

# ==============================
# 4. 결측 제거
# - top3_labels
# - sellerTemperature
# ==============================
before_dropna = len(df)  # 결측 제거 전 row 수

df = df.dropna(
    subset=['top3_labels', 'sellerTemperature']
).copy()  # 결측 제거 후 데이터프레임 복사

after_dropna = len(df)  # 결측 제거 후 row 수
print('결측 제거 후 shape:', df.shape)
print('결측 제거된 row 수:', before_dropna - after_dropna)

# ==============================
# 5. sellerTemperature 숫자형 변환
# ==============================
df['sellerTemperature'] = pd.to_numeric(df['sellerTemperature'], errors='coerce')

# sellerTemperature 변환 후 혹시 생긴 NaN 제거
before_temp = len(df)  # sellerTemperature 숫자 변환 전 row 수
df = df.dropna(subset=['sellerTemperature']).copy()
after_temp = len(df)  # sellerTemperature 숫자 변환 후 row 수
print('sellerTemperature 숫자 변환 후 shape:', df.shape)

print('sellerTemperature 숫자 변환 후 추가 제거 row 수:', before_temp - after_temp)

# ==============================
# 6. price 이상치 제거
# - 2000원 미만 제거
# - 500만원 초과 제거
# ==============================
before_price = len(df)  # price 이상치 제거 전 row 수

df = df[(df['price'] >= 2000) & (df['price'] <= 5_000_000)].copy()

after_price = len(df)
print('price 이상치 제거 후 shape:', df.shape)
print('price 이상치로 제거된 row 수:', before_price - after_price)

# ==============================
# 7. 최종 결과 저장
# ==============================
df.to_csv(output_csv_path, index=False, encoding='utf-8-sig')

print('최종 저장 완료:', output_csv_path)
print('최종 shape:', df.shape)
```
